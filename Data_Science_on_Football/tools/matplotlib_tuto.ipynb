```python
from matplotlib.patches import Arc

# Création de la figure 
fig, ax = plt.subplots(1,1)

# Bordures du stade 
ax.plot([0,0],[0,90], color='black')
ax.plot([0,130], [90,90], color='black')
ax.plot([130,130],[90,0], color='black')
ax.plot([0,130],[0,0], color='black')
ax.plot([65,65],[0,90], color='black')

# Zone de penalty
plt.plot([16.5,16.5],[65,25],color="black")
plt.plot([0,16.5],[65,65],color="black")
plt.plot([16.5,0],[25,25],color="black")
plt.plot([130,113.5],[65,65],color="black")
plt.plot([113.5,113.5],[65,25],color="black")
plt.plot([113.5,130],[25,25],color="black")

# Zone des 6-mètres
plt.plot([0,5.5],[54,54],color="black")
plt.plot([5.5,5.5],[54,36],color="black")
plt.plot([5.5,0.5],[36,36],color="black")
plt.plot([130,124.5],[54,54],color="black")
plt.plot([124.5,124.5],[54,36],color="black")
plt.plot([124.5,130],[36,36],color="black")

# Tracer du rond central
cercle = plt.Circle((65,45),9.15,fill=False)
centre_cercle = plt.Circle((65,45),0.8,fill=False)
point_penalty_gauche = plt.Circle((11,45),0.8,color="black")
point_penalty_droit = plt.Circle((119,45),0.8,color="black")
ax.add_patch(centre_cercle)
ax.add_patch(cercle)
ax.add_patch(point_penalty_gauche)
ax.add_patch(point_penalty_droit)

# Les Arcs 
arc_gauche = Arc((11,45),height=18.3,width=18.3,angle=0,theta1=310,theta2=50,color="black")
arc_droit = Arc((119,45),height=18.3,width=18.3,angle=0,theta1=130,theta2=230,color="black")
ax.add_patch(arc_gauche)
ax.add_patch(arc_droit)

# Affichage
plt.show()
```

```python
def creation_terrain():
    # Création de la figure 
    fig, ax = plt.subplots(1,1)

    # Bordures du stade 
    ax.plot([0,0],[0,90], color='black')
    ax.plot([0,130], [90,90], color='black')
    ax.plot([130,130],[90,0], color='black')
    ax.plot([0,130],[0,0], color='black')
    ax.plot([65,65],[0,90], color='black')

    # Zone de penalty
    plt.plot([16.5,16.5],[65,25],color="black")
    plt.plot([0,16.5],[65,65],color="black")
    plt.plot([16.5,0],[25,25],color="black")
    plt.plot([130,113.5],[65,65],color="black")
    plt.plot([113.5,113.5],[65,25],color="black")
    plt.plot([113.5,130],[25,25],color="black")

    # Zone des 6-mètres
    plt.plot([0,5.5],[54,54],color="black")
    plt.plot([5.5,5.5],[54,36],color="black")
    plt.plot([5.5,0.5],[36,36],color="black")
    plt.plot([130,124.5],[54,54],color="black")
    plt.plot([124.5,124.5],[54,36],color="black")
    plt.plot([124.5,130],[36,36],color="black")

    # Tracer du rond central
    cercle = plt.Circle((65,45),9.15,fill=False)
    centre_cercle = plt.Circle((65,45),0.8,color="black")
    point_penalty_gauche = plt.Circle((11,45),0.8,color="black")
    point_penalty_droit = plt.Circle((119,45),0.8,color="black")
    ax.add_patch(centre_cercle)
    ax.add_patch(cercle)
    ax.add_patch(point_penalty_gauche)
    ax.add_patch(point_penalty_droit)

    # Les Arcs 
    arc_gauche = Arc((11,45),height=18.3,width=18.3,angle=0,theta1=310,theta2=50,color="black")
    arc_droit = Arc((119,45),height=18.3,width=18.3,angle=0,theta1=130,theta2=230,color="black")
    ax.add_patch(arc_gauche)
    ax.add_patch(arc_droit)
    
    # Un peu design pour le terrain 
    ax.set_facecolor('mediumseagreen')
    return fig, ax
```
